```python
# Gambar sebuah jari jari dengan jari jari 1 pada sumbu xy dengan warna hijau, dan daerah luar lingkaran berwarna biru

import matplotlib.pyplot as plt
import numpy as np

circle = plt.Circle((0, 0), 1, color='g')
fig, ax = plt.subplots() 
ax.add_artist(circle)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_facecolor('b')
plt.show()
```
