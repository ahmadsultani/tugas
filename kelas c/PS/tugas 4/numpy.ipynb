```python
import numpy

# Buat array kosong array1 dengan ukuran 2 x 3
array1 = numpy.empty([2,3])

# Buat array array2 dengan ukuran 100 x 150 dengan nilai 0
array2 = numpy.zeros([100,150])

# Buat array array3 dengan ukuran 150 x 100 dengan nilai 1
array3 = numpy.ones([150, 100])

# Tampilkan tipe array1, array2, dan array3
print("Tipe array1:", type(array1))
print("Tipe array2:", type(array2))
print("Tipe array3:", type(array3))

# Tampilkan ukuran array2
print(array2.shape)

# Ubah ukuran array3 menjadi 100 x 150
array3 = array3.reshape(100, 150)

# Gabungkan array2 dan array3 menjadi array4
array4 = numpy.concatenate((array2, array3), axis=1)
print("Ukuran array4:", array4.shape)

# Deklarasikan array kosong array5 untuk menampung nilai float
array5 = numpy.empty([5, 7], dtype=float)
print("Ukuran array5:", array5.shape)
print("Tipe elemen array5:", type(array5[0,0]))

# Gunakan Loop untuk mendisplay setiap baris array2
for baris in array2:
  isiBaris = ""
  for kolom in baris:
    isiBaris += f"{kolom:g} "
  print("[ "+isiBaris+"]") 

# Shallow copy array1
array6 = array1.view()

# Deep copy array1
array7 = array1.copy()

# Perbedaan shallow copy dan deep copy
array1[0,0] = 100
print("array1[0,0] =", array1[0,0]) 
print("array6[0,0] =", array6[0,0])
print("array7[0,0] =", array7[0,0])
print("Saat array1 diubah, array6 ikut berubah (shallow copy), namun array7 tidak berubah (deep copy).") 

```
